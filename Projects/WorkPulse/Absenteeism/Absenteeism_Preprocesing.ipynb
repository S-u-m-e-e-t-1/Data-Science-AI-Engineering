```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age','Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
df
```

```python
df['Reason_1'] = df['Reason_1'].astype(int)
df['Reason_2'] = df['Reason_2'].astype(int)
df['Reason_3'] = df['Reason_3'].astype(int)
df['Reason_4'] = df['Reason_4'].astype(int)
df
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age','Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df
```

```python
df_reason_mod = df.copy()
df_reason_mod
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod
```

```python
df_reason_mod['Date'][0].month
```

```python
df_reason_mod['Month Value'] = df_reason_mod['Date'].dt.month
df_reason_mod
```

```python
column_names_reordered_final = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age','Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
df_reason_mod = df_reason_mod[column_names_reordered_final]
df_reason_mod
```
